```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))

df['y'].value_counts().plot(kind='bar')

plt.title("Subscription Distribution")
plt.xlabel("Subscription")
plt.ylabel("Number of Customers")

plt.show()
```

```python
df['age_group'] = pd.cut(
    df['age'],
    bins=[18, 30, 40, 50, 60, 100],
    labels=['18-30', '31-40', '41-50', '51-60', '60+']
)
```

```python
age_conversion = (
    df.groupby('age_group')['y']
      .value_counts(normalize=True)
      .unstack() * 100
)

age_conversion
```

```python
age_conversion['yes'].plot(kind='bar', color='steelblue',figsize=(7,5))

plt.title("Conversion Rate by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Conversion Rate (%)")
plt.xticks(rotation=0)
plt.show()
```

```python
job_conversion = (
    df.groupby('job')['y']
      .value_counts(normalize=True)
      .unstack() * 100
)

job_conversion.sort_values(by='yes', ascending=False)
```

```python
job_conversion['yes'].sort_values().plot(kind='barh', figsize=(10,6))

plt.title("Conversion Rate by Job")
plt.xlabel("Conversion Rate (%)")
plt.ylabel("Job")
plt.show()
```
